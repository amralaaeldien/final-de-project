```python
import os
import posixpath
from typing import Iterator

import dlt
from dlt.sources import TDataItems

from dlt.sources.filesystem import FileItemDict, filesystem, readers, read_csv  # type: ignore
from pandas import pandas as pd
```

```python
def read_csv_with_duckdb() -> None:
    pipeline = dlt.pipeline(
        pipeline_name="standard_filesystem",
        destination='bigquery',
        dataset_name="final_project_dataset",
    )

    # load all the CSV data, excluding headers
    met_files = readers(
        bucket_url="gs://final-project-ed-zoomcamp", file_glob="saudi_cities_geocoding.csv"
    ).read_csv()

    met_files = pd.DataFrame(met_files)
    
    
    load_info = pipeline.run(met_files, table_name='saudi')

    #data_set = pipeline.dataset.df()
    #data_set.date = pd.to_datetime(data_set.date).date()

    #load_info = pipeline.run(data_set, table_name='saudi_data_set')
    
    print(load_info)
    print(pipeline.last_trace.last_normalize_info)
    
    met_files = readers(
        bucket_url="gs://final-project-ed-zoomcamp", file_glob="DatasetForCoffeeSales2.csv"
    ).read_csv()
    
    met_files = pd.DataFrame(met_files)
    print(met_files)
    met_files.Date = pd.to_datetime(met_files.Date).dt.date
    
    
    load_info = pipeline.run(met_files, table_name="coffee")

    #data_set = pipeline.dataset.df()
    #data_set.date = pd.to_datetime(data_set.date).date()

    #load_info = pipeline.run(data_set, table_name='coffee_data_set')
    
    print(load_info)
    print(pipeline.last_trace.last_normalize_info)
```

```python
read_csv_with_duckdb()
```
